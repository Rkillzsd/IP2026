## Function

```python
def myfun(x):
    return x*3
```

```python
def apply(q, x):
    return q(x)
```

```python
apply(myfun, 7)
```

```python
apply(lambda z: z * 4, 7)
## 간단한 수식을 전개할때는 함수를 선언하기보다, 람다가 더 유용함
```

```python
def myfun(b, c=3, d="hello"):
    return b + c
print(myfun(5,3,"hello"))
print(myfun(5,3))
print(myfun(5))
```

```python
def myfun (a, b, c):
    return a-b
myfun(2, 1, 43)
```

```python
myfun(c=43, b=1, a=2)
```

```python
myfun(2, c=43, b=1)
```
