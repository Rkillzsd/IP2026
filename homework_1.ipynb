```python
import cv2
import numpy as np
def nothing(x):
    pass
drawing = False
ix, iy = -1, -1
cx, cy = -1, -1
org_img = cv2.imread('tree.jpg')
img = org_img.copy()

def draw_circle(event, x, y, flags, param):
    global ix, iy, cx, cy, drawing, img
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
    elif event == cv2.EVENT_MOUSEMOVE:
        cx, cy = x, y
    elif event == cv2.EVENT_LBUTTONUP:
        ix, iy = -1, -1
        cx, cy = -1, -1
        drawing = False

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_circle)
cv2.createTrackbar('value','image',0,255,nothing)
while True:
    img = org_img.copy()
    value = cv2.getTrackbarPos('value', 'image')
    if drawing == True:
        x1, x2 = (ix, cx) if ix < cx else (cx, ix)
        y1, y2 = (iy, cy) if iy < cy else (cy, iy)
        if x1 != x2 and y1 != y2:
            img[y1:y2, x1:x2, 2] = 255
    cv2.putText(img,f'Mouse Position : ({ix}, {iy}) - ({cx}, {cy})-{value}', (10,30), font, 0.7,(255,255,255),2,cv2.LINE_AA)
    cv2.imshow('image', img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
```
