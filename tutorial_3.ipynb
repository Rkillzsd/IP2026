## Dictionaries: Like maps in Java

```python
d = {'user': 'bozo', 'pswd': 1234}
print(d['user'])
print(d['pswd'])
print(d['bozo']) ## 이건 안됨 에러뜸
```

```python
d['user'] = 'clown' ## 요소의 값 변경
d
```

```python
d['id'] = 45 ## 요소 추가
d
```

```python
d = {'user':'bozo', 'p':1234, 'i':34}
del d['user'] ## 삭제
d
```

```python
d.clear() ## 딕셔너리의 모든 요소 삭제
d
```

```python
a=[1,2]
del a[1] ## 리스트에서도 동작함
a
```

```python
d = {'user':'bozo', 'p':1234, 'i':34} 
d.keys() ## 딕셔너리의 키를 리스트로 반환
d.values() ## 딕셔너리의 값을 리스트로 반환
d.items() ## 딕셔너리의 요소를 (키, 값) 튜플로 반환
```
