## List

```python
li = ['abc', 23, 4.34, 23]
li[1] = 45 
print (li)
```

```python
## 튜플
t = (23, 'abc', 4.56, (2,3), 'def')
t[2] = 3.14
print (t)
## 에러가 나옴
```

```python
li = [1, 11, 3, 4, 5]
li.append('a') ## 리스트의 맨 뒤에 요소 추가
print (li)
li.insert(2, 'i') ## 리스트의 2번 인덱스에 'i' 추가
print (li)
## 중간고사에 나올 수 있음

```

```python
li.extend([9, 8, 7]) ## extend는 리스트의 각 요소를 하나씩 추가
print (li)

li.append([10, 11, 12]) ## append는 리스트 전체를 하나의 요소로 추가
print (li)
```

```python
li = ['a', 'b', 'c', 'b']
li.index('b') ## 리스트에서 'b'의 인덱스 반환
```

```python
li.count('b') ## 리스트에서 'b'의 개수 반환
```

```python
li.remove('b') ## 리스트에서 'b'를 제거
print (li)
```

```python
li = [5, 2, 6, 8]
li.reverse() ## 리스트의 요소 순서를 뒤집음
print (li)
```

```python
li.sort() ## 리스트의 요소를 오름차순으로 정렬
print (li)
```

```python
li.sort(some_function) ## [some_function]을 기준으로 리스트의 요소를 정렬
```

리스트는 튜플보다 속도는 느리지만 더 강력한 기능을 제공합니다. 
리스트는 수정이 가능하며, 다양한 편리한 연산을 수행할 수 있습니다. 
튜플은 불변하며 기능이 더 적습니다. 
튜플과 리스트 간 변환에는 `list()` 및 `tuple()` 함수를 사용합니다. 
예: `li = list(tu)` `tu = tuple(li)`
